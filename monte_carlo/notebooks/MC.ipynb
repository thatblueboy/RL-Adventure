```python
import os
import sys
import gymnasium as gym
sys.path.append('/home/thatblueboy/ML/RL/adventure/_environments')
sys.path.append('/home/thatblueboy/ML/RL/adventure/monte_carlo/src')
import frozen_lake_custom
import MC
```

```python
trainEnv = gym.make('FrozenLake-v1',map_name = '4x4', render_mode = 'ansi', is_slippery=False)
# trainEnv = FrozenLakeCustomEnv(map_name='4x4', is_slippery=False)

MCAgent = MC.MCAgent(trainEnv)
MCAgent.train_policy(5000)
```

```python
def test(env, agent, num):
    for i in range(num):
        state = env.reset()[0]
        for i in range(100):
            action = agent.deterministicPolicy(state) 
            next_state, reward, terminated, truncated, info = env.step(action)
            state = next_state
            if terminated or truncated:
                break
    env.close()
```

```python
testEnv = gym.make('FrozenLake-v1',map_name = '4x4', render_mode = 'human', is_slippery=False)
test(testEnv, MCAgent, 1)
```
